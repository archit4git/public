Records with B2C Uber ID (prop70) | June 2022

```python
%%read_sql -c QS_CONNECTION
select
    count(_id) recordsWithB2cUberId
from 
    global_prod_midvalues
where 
    from_utc_timestamp(timestamp, 'America/New_York') >= to_timestamp('2022-06-01 00:00:00')
    and from_utc_timestamp(timestamp, 'America/New_York') < to_timestamp('2022-06-30 00:00:00')
    and _experience.analytics.customDimensions.props.prop70 is not null
```

Records with B2C Uber ID (prop70) by Date | June 2022

```python
%%read_sql -c QS_CONNECTION
select
    date_trunc('DD', timestamp) dateOfEvent,
    count(_id) recordsWithB2cUberId
from 
    global_prod_midvalues
where 
    from_utc_timestamp(timestamp, 'America/New_York') >= to_timestamp('2022-06-01 00:00:00')
    and from_utc_timestamp(timestamp, 'America/New_York') < to_timestamp('2022-06-30 00:00:00')
    and _experience.analytics.customDimensions.props.prop70 is not null
group by 1
order by 1
```
